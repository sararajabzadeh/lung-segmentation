```python
import torch
from lungseg.models import UNet

model = UNet()
x = torch.randn(1, 1, 256, 256)

for i, down in enumerate(model.downs):
    x = down(x)
    print(f"level {i} after DoubleConv:", tuple(x.shape))
    x = model.pool(x)
    print(f"level {i} after pool:     ", tuple(x.shape))
```

```python
model = UNet(in_channels=1, out_channels=1)
x = torch.randn(1, 1, 256, 256)
out = model(x)
print(out.shape)
assert out.shape == x.shape
```
